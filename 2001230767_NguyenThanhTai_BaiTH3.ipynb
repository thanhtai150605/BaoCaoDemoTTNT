Minimax (Ng∆∞·ªùi ch∆°i t·ª± nh·∫≠p)

Quy ∆∞·ªõc

B·∫°n: X

M√°y: O

Nh·∫≠p t·ªça ƒë·ªô d·∫°ng: row col (0‚Äì2)

```python
import math

PLAYER = 'X'
AI = 'O'
EMPTY = ' '

def print_board(board):
    print("\nB√†n c·ªù:")
    for row in board:
        print(" | ".join(row))
    print()

def create_board():
    return [[EMPTY]*3 for _ in range(3)]

def is_winner(board, p):
    for i in range(3):
        if all(board[i][j] == p for j in range(3)) or \
           all(board[j][i] == p for j in range(3)):
            return True
    if all(board[i][i] == p for i in range(3)) or \
       all(board[i][2-i] == p for i in range(3)):
        return True
    return False

def is_full(board):
    return all(board[i][j] != EMPTY for i in range(3) for j in range(3))

def evaluate(board):
    if is_winner(board, PLAYER): return 1
    if is_winner(board, AI): return -1
    return 0

def minimax(board, is_max):
    score = evaluate(board)
    if score != 0 or is_full(board):
        return score

    if is_max:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == EMPTY:
                    board[i][j] = PLAYER
                    best = max(best, minimax(board, False))
                    board[i][j] = EMPTY
        return best
    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == EMPTY:
                    board[i][j] = AI
                    best = min(best, minimax(board, True))
                    board[i][j] = EMPTY
        return best

def best_move_minimax(board):
    best_val = math.inf
    move = None
    for i in range(3):
        for j in range(3):
            if board[i][j] == EMPTY:
                board[i][j] = AI
                val = minimax(board, True)
                board[i][j] = EMPTY
                if val < best_val:
                    best_val = val
                    move = (i, j)
    return move

# ===== GAME LOOP =====
board = create_board()

while True:
    print_board(board)

    # Ng∆∞·ªùi ch∆°i nh·∫≠p
    r, c = map(int, input("B·∫°n nh·∫≠p (row col): ").split())
    if board[r][c] != EMPTY:
        print("√î ƒë√£ c√≥ qu√¢n, nh·∫≠p l·∫°i!")
        continue
    board[r][c] = PLAYER

    if is_winner(board, PLAYER):
        print_board(board)
        print("üéâ B·∫°n th·∫Øng!")
        break

    if is_full(board):
        print_board(board)
        print("ü§ù H√≤a!")
        break

    # M√°y ch∆°i
    ai_move = best_move_minimax(board)
    board[ai_move[0]][ai_move[1]] = AI
    print("M√°y ƒëi:", ai_move)

    if is_winner(board, AI):
        print_board(board)
        print("üíª M√°y th·∫Øng!")
        break

```

V·ªõi b√†n c·ªù 3√ó3, Minimax duy·ªát ƒë∆∞·ª£c to√†n b·ªô kh√¥ng gian tr·∫°ng th√°i n√™n m√°y kh√¥ng bao gi·ªù thua n·∫øu ch∆°i ƒë√∫ng.

Tuy nhi√™n, Minimax c√≥ nh∆∞·ª£c ƒëi·ªÉm l√† t·ªën th·ªùi gian t√≠nh to√°n khi kh√¥ng gian t√¨m ki·∫øm l·ªõn, do ƒë√≥ trong c√°c b√†i to√°n ph·ª©c t·∫°p h∆°n c·∫ßn s·ª≠ d·ª•ng Alpha‚ÄìBeta pruning ƒë·ªÉ t·ªëi ∆∞u.

C√†i ƒë·∫∑t thu·∫≠t to√°n c·∫Øt t·ªâa Alpha‚ÄìBeta cho b√†i to√°n Tic-Tac-Toe (3√ó3)

```python
import math

PLAYER = 'X'
AI = 'O'
EMPTY = ' '

# ================== T·∫†O V√Ä HI·ªÇN TH·ªä B√ÄN C·ªú ==================

def create_board():
    return [[EMPTY]*3 for _ in range(3)]

def print_board(board):
    print("\nB√†n c·ªù:")
    for row in board:
        print(" | ".join(row))
    print()

# ================== KI·ªÇM TRA TR·∫†NG TH√ÅI ==================

def is_winner(board, p):
    for i in range(3):
        if all(board[i][j] == p for j in range(3)) or \
           all(board[j][i] == p for j in range(3)):
            return True
    if all(board[i][i] == p for i in range(3)) or \
       all(board[i][2-i] == p for i in range(3)):
        return True
    return False

def is_full(board):
    return all(board[i][j] != EMPTY for i in range(3) for j in range(3))

def evaluate(board):
    if is_winner(board, PLAYER):
        return 1
    if is_winner(board, AI):
        return -1
    return 0

# ================== THU·∫¨T TO√ÅN ALPHA-BETA ==================

def alphabeta(board, alpha, beta, is_max):
    score = evaluate(board)

    if score != 0 or is_full(board):
        return score

    if is_max:  # l∆∞·ª£t ng∆∞·ªùi ch∆°i
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == EMPTY:
                    board[i][j] = PLAYER
                    best = max(best, alphabeta(board, alpha, beta, False))
                    board[i][j] = EMPTY
                    alpha = max(alpha, best)
                    if beta <= alpha:
                        break
        return best
    else:  # l∆∞·ª£t m√°y
        best = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == EMPTY:
                    board[i][j] = AI
                    best = min(best, alphabeta(board, alpha, beta, True))
                    board[i][j] = EMPTY
                    beta = min(beta, best)
                    if beta <= alpha:
                        break
        return best

def best_move_alphabeta(board):
    best_val = math.inf
    move = None

    for i in range(3):
        for j in range(3):
            if board[i][j] == EMPTY:
                board[i][j] = AI
                val = alphabeta(board, -math.inf, math.inf, True)
                board[i][j] = EMPTY
                if val < best_val:
                    best_val = val
                    move = (i, j)

    return move

# ================== V√íNG L·∫∂P TR√í CH∆†I ==================

board = create_board()

while True:
    print_board(board)

    # Ng∆∞·ªùi ch∆°i nh·∫≠p n∆∞·ªõc ƒëi
    try:
        r, c = map(int, input("Nh·∫≠p n∆∞·ªõc ƒëi (row col): ").split())
        if board[r][c] != EMPTY:
            print("√î ƒë√£ c√≥ qu√¢n, nh·∫≠p l·∫°i!")
            continue
    except:
        print("Nh·∫≠p sai ƒë·ªãnh d·∫°ng!")
        continue

    board[r][c] = PLAYER

    if is_winner(board, PLAYER):
        print_board(board)
        print("üéâ B·∫°n th·∫Øng!")
        break

    if is_full(board):
        print_board(board)
        print("ü§ù H√≤a!")
        break

    # M√°y ƒëi
    ai_move = best_move_alphabeta(board)
    board[ai_move[0]][ai_move[1]] = AI
    print("üíª M√°y ƒëi:", ai_move)

    if is_winner(board, AI):
        print_board(board)
        print("‚ùå M√°y th·∫Øng!")
        break

```
